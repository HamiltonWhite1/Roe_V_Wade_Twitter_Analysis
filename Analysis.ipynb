```python
import pandas as pd
```

```python
cumulative_tweets = pd.read_json('Analysis_Data/Analysis.json')
```

```python
#Models for politically-self-identifying users. This should grow over time, resulting in increasingly accurate
#findings

democrat = ['ginsburg', 'rbg', 'pro-choice', 'liberal', 'democrat', 'biden supporter', 'dem',
            'progressive', 'republicans are killing democracy', '#LGQBTQ', 'anti-maga', 'cult',
            'against the gop', 'pronouns', '#blacklivesmatter', '#blm']

republican = ['pro-life', 'republican', 'conservative', 'christian', 'catholic', 'trump', 'god',
              'my country', 'presbyterian', 'patriot', 'white', 'straight', 'president trump',
              '#alllivesmatter']
```

```python
#So far I have coded a test below to identify users who identify as democrat and users who identify as republican
#tomorrow work on continuing the development of the NLP model for identifying those users, as well as 
#engineering a body of code to evaluate the body of the tweets of the user

total_tweets = len(cumulative_tweets)
twitter_user = 0
test_dems = 0
test_repubs = 0
while twitter_user < len(cumulative_tweets):
    tweet = cumulative_tweets['user'][twitter_user]['description'].split()
    for word in tweet:
        if word in democrat:
            test_dems += 1
    for word in tweet:
        if word in republican:
            test_repubs += 1
    twitter_user += 1
print(f"There are {test_dems} self-identified democrats and {test_repubs} self-identified republicans out of {total_tweets} unique tweets")
    
    
```
